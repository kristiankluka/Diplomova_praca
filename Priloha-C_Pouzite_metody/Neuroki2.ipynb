```python
import neurokit2 as nk
import pandas as pd
features = pd.read_csv('Hnev_EKG.csv', sep=';')
#features = data.drop(['ID'], axis=1)
#target = data['Label']
#row=features.iloc[4]

features
```

```python
features.iloc[4]
```

```python
df_new = pd.DataFrame()
for index, row in features.iterrows():
  processed_data, info = nk.ecg_process(row, sampling_rate=1)
  results = nk.bio_analyze(processed_data, sampling_rate=1)
  df_new = df_new.append(results, ignore_index=True)
  #print("1")
```

```python
df_new
```

```python
pd.set_option('max_rows', 99999)
pd.set_option('max_colwidth', 400)
pd.describe_option('max_colwidth')
```

```python
df_new = df_new.drop(['HRV_SDANN2','HRV_SDNNI2','HRV_SDANN5','HRV_SDNNI5','HRV_ULF'], axis=1)
df_new = df_new.fillna(0)
df_new.isna().sum()
```

```python
df_new.to_excel("EKG_hnev.xlsx")  
```
